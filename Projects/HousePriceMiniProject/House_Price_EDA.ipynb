## Data Cleaning

```python
# Fill missing 'Price (in rupees)' with the mean
data['Price (in rupees)'].fillna(data['Price (in rupees)'].mean(), inplace=True)

```

```python
# Fill missing 'Price (in rupees)' with the mean
data['Price (in rupees)'].fillna(data['Price (in rupees)'].mean(), inplace=True)
```

```python
# Fill missing 'Carpet Area' with median (safer than mean for skewed data)
data['Carpet Area'] = data['Carpet Area'].astype(str)
data['Carpet Area'] = data['Carpet Area'].str.extract('(\d+)')
data['Carpet Area'] = pd.to_numeric(data['Carpet Area'], errors='coerce')
data['Carpet Area'] = data['Carpet Area'].fillna(data['Carpet Area'].median())
```

```python

# Fill missing 'Status' with 'Unknown'
data['Status'].fillna('Unknown', inplace=True)
```

```python
# Drop columns with mostly missing values (>80%)
threshold = 0.8 * len(data)
cols_to_drop = [col for col in data.columns if data[col].isnull().sum() > threshold]
data.drop(columns=cols_to_drop, inplace=True)
```

## Aggregations & Relationships

```python
# Average price
average_price = data['Price (in rupees)'].mean()
print(f"Average Price: {average_price:.2f} rupees")
```

```python
# Average price by Location
average_price_by_location = data.groupby('location')['Price (in rupees)'].mean().sort_values(ascending=False)
print("Average Price by Location:" , average_price_by_location)
```

```python
# Min and Max carpet area
min_area = data['Carpet Area'].min()
max_area = data['Carpet Area'].max()
print(f"Minimum Carpet Area: {min_area} sqft")
print(f"Maximum Carpet Area: {max_area} sqft")
```

```python

# Average Price by Status
average_price_by_status = data.groupby('Status')['Price (in rupees)'].mean().sort_values(ascending=False)
print("Average Price by Status:" , average_price_by_status)
```

```python
# Average Carpet Area by Location
average_area_by_location = data.groupby('location')['Carpet Area'].mean().sort_values(ascending=False)
print("Average Carpet Area by Location:" , average_area_by_location)
```

## General Questions on Data

```python
# Display the first 10 rows of the cleaned dataset.
data.head(10)
```

```python

# Count the number of flats per Location.
flats_per_location = data['location'].value_counts()
print("Number of Flats per Location:\n", flats_per_location)
```

```python
# Find the top 5 most expensive flats
top_5_expensive_flats = data.nlargest(5, 'Price (in rupees)')
print("Top 5 Most Expensive Flats:\n", top_5_expensive_flats[['Title', 'Price (in rupees)', 'location']])
```

```python
# Filter all flats where Status = Ready to Move
ready_to_move_flats = data[data['Status'] == 'Ready to Move']
print("Flats with Status 'Ready to Move':\n", ready_to_move_flats[['Title', 'Price (in rupees)', 'Carpet Area', 'location']])
```

```python
# Calculate median Price and compare with mean Price
median_price = data['Price (in rupees)'].median()
mean_price = data['Price (in rupees)'].mean()
print(f"Median Price: {median_price:.2f} rupees")
print(f"Mean Price: {mean_price:.2f} rupees")
```

```python
# Explore relationship between Floor and Price
floor_price_relationship = data.groupby('Floor')['Price (in rupees)'].mean().sort_values(ascending=False)
print("Average Price by Floor:\n", floor_price_relationship)
```
